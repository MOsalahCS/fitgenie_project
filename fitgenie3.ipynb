```python
import pandas as pd 
```

```python
data = pd.read_csv('user_data_new4.csv')

data 
```

```python
data = data.drop(['Unnamed: 0.1','Unnamed: 0'],axis=1)

data 
```

```python
cat_feat = ["Gender","Allergies","Fitness Goals","Work Out Level"]
cont_feat = [x for x in data.columns.tolist() if x not in cat_feat]
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
cat_data_org=data[cat_feat]
```

```python
cat_data_org
```

```python
cat_data = cat_data_org.copy()
```

```python
labels_encoder = {}

# Iterate through each column and encode categorical values
for column in cat_feat:
    label_encoder = LabelEncoder()
    label_encoder.fit(cat_data_org[column])
    cat_data[column] = label_encoder.transform(cat_data_org[column])
    labels_encoder[column] = label_encoder
```

```python
labels_encoder
```

```python
cat_data
```

```python
scaler = StandardScaler()
cont_data=scaler.fit_transform(data[cont_feat])
```

```python
cont_data = pd.DataFrame(cont_data,columns=cont_feat)
```

```python
cont_data
```

```python
data_prep=pd.concat([cont_data,cat_data],axis=1)
```

```python
data_prep
```

```python
from kmodes.kprototypes import KPrototypes
```

```python
test_model = KPrototypes(n_clusters=20, init='Huang', random_state=42)
test_model.fit_predict(data_prep, categorical=[9,10,11,12])
```

```python
clusters = test_model.predict(data_prep, categorical=[9,10,11,12])
```

```python
clusters
```

```python
data_prep["clusters"]=clusters
```

```python
data_prep
```

```python
from sklearn.manifold import TSNE 
import plotly.express as px
import plotly.graph_objects as go 
```

```python
def plot_pca_2d(df, title = "PCA Space", opacity=0.8, width_line = 0.1):

    df = df.astype({"cluster": "object"})
    df = df.sort_values("cluster")


    fig = px.scatter(
          df, 
          x='comp1', 
          y='comp2', 
          color='cluster',
          
          # symbol = "cluster",
          
          color_discrete_sequence=px.colors.qualitative.Vivid,
          title=title).update_traces(
              # mode = 'markers',
              marker={
                  "size": 3,
                  "opacity": opacity,
                  # "symbol" : "diamond",
                  "line": {
                      "width": width_line,
                      "color": "black",
                  }
              }
          ).update_layout(
                  width = 800, 
                  height = 800, 
                  autosize = True, 
                  showlegend = True,
                  legend=dict(title_font_family="Times New Roman",
                              font=dict(size= 20)),
                  scene = dict(xaxis=dict(title = 'comp1', titlefont_color = 'black'),
                              yaxis=dict(title = 'comp2', titlefont_color = 'black')),
                              
                  font = dict(family = "Gilroy", color  = 'black', size = 20))
        
    
    fig.show()
```

```python
sampling_data = data_prep.sample(frac=0.1, replace=True, random_state=1)

sampling_data = data_prep.copy()


df_tsne_2d = TSNE(
                  n_components=2, 
                   
                  init='pca',  
                  n_iter = 850).fit_transform(sampling_data)

df_tsne_2d = pd.DataFrame(df_tsne_2d, columns=["comp1", "comp2"])
df_tsne_2d["cluster"] = data_prep["clusters"]
plot_pca_2d(df_tsne_2d, title = "t-sne Space")
```

```python
cat_agg={col: lambda x: x.value_counts().sort_values(ascending=False).index[0] for col in cat_feat}
cat_agg
```

```python
cont_agg={col: 'mean' for col in cont_feat}
cont_agg
```

```python
all_agg = {**cat_agg , **cont_agg}
all_agg
```

```python
cls_agg=data_prep.groupby('clusters').agg(
    all_agg
).reset_index()
```

```python
cls_agg
```

```python
cls_agg_decoded = cls_agg.copy()
```

```python

for col in cls_agg[cat_feat].columns:
    
    cls_agg_decoded[col]=labels_encoder[col].inverse_transform(cls_agg[col]) 
```

```python
cls_agg_decoded
```

```python
cls_agg_decoded[cont_feat]=scaler.inverse_transform(cls_agg[cont_feat]) 
```

```python
res = []
for column in data_prep[cat_feat+['clusters']].columns.difference(['clusters']):
    # Group by 'Cluster' column and count occurrences of each value in the current column
    result = data_prep[cat_feat+['clusters']].groupby('clusters')[column].value_counts().unstack(fill_value=0)
    result.columns = [column+" "+str(x) for x in result.columns.tolist() ]
    res.append(result)
```

```python
pd.concat(res,axis=1)
```

```python
for cluster, group in data_prep[cat_feat+['clusters']].groupby('clusters'):
    print(f"Cluster {cluster}:")
    # Iterate over columns (excluding the 'Cluster' column)
    for column in data_prep[cat_feat+['clusters']].columns.difference(['clusters']):
        # Count occurrences of each value in the column
        value_counts = group[column].value_counts()
        print(f"\t{column}:")
        print(value_counts)
        print()  # Add a newline for readability
    print()  # Add an additional newline to separate clusters
```
